# Tahap 2 : 1D Visualization
* Membuat Bar Chart dari variabel categorical 
* Membuat Histogram dari variabel numerical

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
house_categorical = pd.read_csv('house_categorical.csv',index_col=0)
house_numerical = pd.read_csv('house_numeric.csv',index_col=0)
```

### Categorical Variable Visualization
* Untuk keperluan analisis variabel kategori kita akan gunakan 3 contoh variabel yaitu Utilities,HouseStyle dan ExterQual

### Utilities

```python
sns.barplot(x=house_categorical['Utilities'].value_counts().index,y=house_categorical['Utilities'].value_counts(normalize = True))
plt.title ('Bar Chart')
plt.xlabel ('Utilities')
plt.ylabel ('Proportion')
plt.show()
```

Analisa:
* Variabel Utilities terdiri dari kategori fasilitas yang dimiliki setiap properti.
* Hampir 100% data properti adalah dengan kategori AllPub  
* AllPub(All Public Utilities) adalah kategori properti yang menyediakan fasilitas lengkap seperti listrik, gas, air dan septic tank
* NoSeWa(No Septic Tank and Water) adalah kategori properti yang hanya menyediakan fasilitas listrik dan gas saja

### HouseStyle

```python
sns.barplot(x=house_categorical['HouseStyle'].value_counts().index,y=house_categorical['HouseStyle'].value_counts(normalize = True))
plt.title ('Bar Chart')
plt.xlabel ('HouseStyle')
plt.ylabel ('Proportion')
plt.show()
```

Analisa:
* Variabel HousStyle terdiri dari gaya/desain bangunan
* Data berdasarkan HouseStyle terlihat imbalanced
* Hampir 50% data adalah properti dengan desain 1Story (1lantai)
* 30% data properti dengan desain 2Story (2 lantai)
* 10% data properti dengan desain 1.5Fin (1.5 lantai)
* desain lainnya seperti: Slvl, SFoyer, 1.5Unf, 2.5Unf dan 2.5Fin memiliki data <5%

```python
house_categorical['HouseStyle'].value_counts()
```

### ExterQual

```python
sns.barplot(x=house_categorical['ExterQual'].value_counts().index,y=house_categorical['ExterQual'].value_counts(normalize = True))
plt.title ('Bar Chart')
plt.xlabel ('ExterQual')
plt.ylabel ('Proportion')
plt.show()
```

Analisa:
* Variable ExterQual terdiri dari kategori kualitas bahan exterior dari properti
    TA: Typical/Average
    Gd: Good

* Data penjualan berdasarkan ExterQual terlihat imbalanced

* < 5% dengan kualitas sangat bagus dan juga kualitas biasa 

## Numerical Variabel Visualization
* Untuk keperluan analisis kita gunakan contoh variabel yaitu : SalePrice, GrLivArea dan PoolArea

### SalePrice

```python
x = house_numerical['SalePrice']
sns.distplot(x, axlabel = x.name)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label= "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.title('Histogram')
plt.show()
```

Analisa:
* Sale Price berarti harga properti yang dijual, berbanding lurus dengan kualitas  propertinya
* Distribusi right-skewed
* Kebanyakan data ada pada range harga 100.000 sampai 200.000


```python
#Transformasi log10 pada variabel SalePrice
x = np.log10(house_numerical['SalePrice'])
sns.distplot(x, axlabel = x.name)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label= "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.title('Histogram')
plt.show()
```

# Tahap 3 : 2D Visualization

 * ### Visualisasi hubungan antara dataset GrLivArea dengan SalePrice

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
house_categorical = pd.read_csv('house_categorical.csv',index_col=0)
house_numerical = pd.read_csv('house_numeric.csv',index_col=0)
```

### Scatterplot


```python
ax = sns.scatterplot(y='SalePrice', x='GrLivArea', data=house_numerical)
plt.title('Linear graph')
plt.ylabel('SalePrice')
plt.xlabel('GrLiveArea')
plt.show()
```

Analisa:
* Dari scatterplot ini terlihat bahwa variabel GrLivArea memiliki korelasi positif dengan variabel SalePrice
* Artinya semakin luas area bangunan di atas permukaan tanah maka semakin tinggi pula harga jual propertinya
* Namun ada sejumlah data yang kemungkinan adalah outlier yaitu titik di ujung kanan
  

### Visualisasi hubungan antara dataset GarageArea dengan SalePrice

### Scatterplot

```python
import pandas as pd
import numpy as np
import seaborn as sns
import seaborn as sns
%matplotlib inline
ax = sns.scatterplot(y='SalePrice', x='GarageArea', data=house_numerical)
plt.title('Linear graph')
plt.ylabel('SalePrice')
plt.xlabel('GarageArea')
plt.show()
```

Analisa:
* Dari scatterplot ini terlihat bahwa variabel GarageArea masih mengandung value 0 yang secara logika seharusnya tidak memiliki SalePrice. Maka perlu melakukan filter data pada kolom GarageArea untuk value > 0

```python
#Filtering GarageArea value > 0
filtered_df = house_numerical[(house_numerical['GarageArea']>0)]
filtered_df
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import seaborn as sns
%matplotlib inline
ax = sns.scatterplot(y='SalePrice', x='GarageArea', data=filtered_df)
plt.title('Linear graph')
plt.ylabel('SalePrice')
plt.xlabel('GarageArea')
plt.show()
```

Analisa :
* Variabel GarageArea yang memiliki value > 0 lebih masuk akal memiliki harga jual (SalePrice)
* Scatterplot ini menunjukkan korelasi positif antara variabel GarageArea dengan SalePrice.
* Artinya semakin luas area garasi yang dimiliki sebuah properti akan menaikkan nilai harga jualnya.
* Namun masih terdapat data anomali yang kemungkinan adalah outlier yaitu titik-titik pada ujung kanan.

```python
#Menghitung nilai korelasi antara variabel GrlivArea dengan SalePrice

r_GrLiveArea_SalePrice = house_numerical['GrLivArea'].corr(house_numerical['SalePrice'])
print(r_GrLiveArea_SalePrice)
```

Analisa :
Nilai Korelasi antara variabel GrLiveArea dengan SalePrice tergolong sedang (cukup kuat) positif karena mendekati 1.
